```python
# ***** imports ******

from selenium import webdriver
from time import sleep
from datetime import datetime
from IPython.display import clear_output

# ***** Setup web driver *****

chrome_options = webdriver.ChromeOptions()
# Desable notifications popup
prefs = {"profile.default_content_setting_values.notifications": 2}
chrome_options.add_experimental_option("prefs", prefs)
# TODO: put your chromedriver's path below
chromedriver_bin_path = 'chromedriver'
# init the web driver
driver = webdriver.Chrome(executable_path=chromedriver_bin_path, options=chrome_options)

# ***** Facebook login *****

# open facebook login page
driver.get("https://www.facebook.com/")
# Search for username and password text fileds
username_field = driver.find_element_by_id("email")
password_field = driver.find_element_by_id("pass")
# login button
login_btn = driver.find_element_by_name('login')
# fill the username and password
# TODO: put your username and password below
user = "put your email or phone number here"
password = "put your password here"
username_field.send_keys(user)
password_field.send_keys(password)
# click login button
login_btn.click()
# Wait for 2 seconds until the login process done
sleep(2)
# go to profile
driver.get("http://m.facebook.com/profile.php")
sleep(3)

# ***** init counters *****

delete_posts = True
deleted_posts = 0
hided_posts = 0
start_time = datetime.now()

# ***** Delete/hide posts *****

while delete_posts:
    clear_output()
    try:
        # post menu (three doted button on the top right of every Facebook post)
        class_name = "_4s19.sec"
        menu_btn = driver.find_element_by_class_name(class_name)
        menu_btn.click()
        sleep(1)
        try:  # to Delete the post if the delete option is available
            print("Post found trying to delete it...")
            # get delete option
            elt = driver.find_element_by_xpath(
                "//a/span/span/div[contains(text(), 'Delete post')]")
            # delete option is not clickable so we need to get its clickable anscestor which is an <a> tag
            parent = elt.find_element_by_xpath(".//ancestor::a")
            # click on the parent element to delete the post
            parent.click()
            # wait for the confirmation button to appair
            sleep(2)
            # get the confirmation button and click it
            hide_btn = driver.find_element_by_xpath(
                "//a[contains(@title, 'Delete')]")
            hide_btn.click()
            sleep(3)
            print("Post deleted.")
            deleted_posts = deleted_posts + 1
        except:  # if not
            print("Couldn't delete the post, tryin to hide it...")
            try:  # to hide the post if this option is available
                # get delete option
                elt = driver.find_element_by_xpath(
                    "//a/span/span/div[contains(text(), 'Hide from profile')]")
                # delete option is not clickable so we need to get its clickable anscestor which is an <a> tag
                parent = elt.find_element_by_xpath(".//ancestor::a")
                # click on the parent element to delete the post
                parent.click()
                # wait for the confirmation button to appair
                sleep(3)
                # get the confirmation button and click it
                hide_btn = driver.find_element_by_xpath(
                    "//a[contains(@title, 'Hide')]")
                hide_btn.click()
                sleep(3)
                print('Post Hided.')
                hided_posts = hided_posts + 1
            except:
                print("Couldn't hide the post")
        print("Deleted posts:", deleted_posts, "| hided posts:", hided_posts,
          "| total:", deleted_posts + hided_posts, "| passed time:",
          datetime.now() - start_time)
        print('Refreshing...')
        driver.refresh()
        sleep(2)
    except:
        print("Nothing to Delete or hide.")
        delete_posts = False
        print("Deleted posts:", deleted_posts, "| hided posts:", hided_posts,
              "| total:", deleted_posts + hided_posts, "| passed time:",
              datetime.now() - start_time)
    
```
